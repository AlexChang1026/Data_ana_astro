```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
mean = 0 
sigma = 10 
input_number = (10**np.arange(1,5,0.5)).astype(int)
var_mc = np.zeros(len(input_number))
var_theo = np.zeros(len(input_number), dtype=np.float64)
```

```python
for y in range(len(input_number)):
    var = np.zeros(1000)
    for x in range(1000):
        normal = np.random.normal(mean,sigma,input_number[y])
        var[x] = np.std(normal)*np.std(normal)
    var_mc[y] = np.std(var)*np.std(var)
    var_theo[y] = (2*(sigma**4)/input_number[y])
```

```python
plt.figure().set_size_inches(7,7)
plt.plot(input_number,var_theo,label=r"Theoretical Prediction")
plt.plot(input_number,var_mc,label=r"MC",marker='o')
plt.xlabel("Input Number",fontsize=18)
plt.ylabel("Variance of sample Variance")
plt.yscale('log')
plt.xscale('log')
plt.legend()
```
