```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n_simulation = 1000
input_number = (10**np.arange(1,5,0.5)).astype(int)
cauchy_sim = np.random.standard_cauchy(n_simulation)
#cauchy_sim = cauchy_sim[(cauchy_sim>-25) & (cauchy_sim<25)]
std_n = []
std_theo = np.zeros(len(input_number))
for n_point in range(len(input_number)): 
    mean_value = np.zeros(n_simulation)
    for i in range(n_simulation):
        random_data = np.random.standard_cauchy(input_number[n_point])
       # random_data = random_data[(random_data>-25) & (random_data<25)]
        mean_value[i] = np.mean(random_data)  
    std_n.append(np.std(mean_value))
    std_theo[n_point] = np.std(cauchy_sim)/np.sqrt(input_number[n_point])
plt.figure().set_size_inches(7,7)    
plt.plot(input_number,std_n,label=r"MC",marker='o')
plt.plot(input_number,std_theo,label=r"Theoretical Prediction")
plt.xlabel("Input Number",fontsize=18)
plt.ylabel("Variance")
plt.legend()
plt.yscale('log')
plt.xscale('log')
```
