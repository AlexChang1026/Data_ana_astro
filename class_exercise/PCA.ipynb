```python
import astropy.io.fits as pf
from sklearn.decomposition import  PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pf.open("clean_spectra.fits")
```

```python
wavelength = 10**data[2].data
```

```python
plt.plot(wavelength,data[0].data[10,:])
plt.xlabel('Wavelength [$\\rm \\AA$]',fontsize=20)
```

```python
pca = PCA(10)
pca.fit(data[0].data)
```

```python
plt.plot(wavelength,pca.mean_)
```

```python
plt.plot(wavelength,pca.components_[0])
```

```python
plt.figure(figsize=(15,20))
for i in range(0,5):
    plt.subplot(5,1,i+1)
    plt.plot(wavelength,pca.components_[i])
```
