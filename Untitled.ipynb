```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
mu = 0 
sigma = 1 
n_simulation = 99
monte_carlo = np.zeros(n_simulation)
n_point = np.arange(100,10000,100)

for i_point in range(0,len(n_point)):
    
    mean_value = np.zeros(i_point)
    
    for i_sim in range(0,i_point):
        random_data2 = np.random.normal(mu,sigma,i_point)
        mean_value[i_sim] = np.mean(random_data2) 
            
            
    std1 = np.std(mean_value)
    monte_carlo[i_point] = std1
    
plt.plot(n_point,monte_carlo)    
```
